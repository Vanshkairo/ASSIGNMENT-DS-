```python
#Q8
df12 = pd.DataFrame({'A':[3,8,6,2,9],
                     'B':[5,2,9,3,1],
                     'C':[1,7,4,5,2]})
df12
```

```python
def filter_dataframe(df12):
    selected_rows = df12[(df12['A'] > 5) & (df12['B'] < 10)]
    return selected_rows
```

```python
selected_df = pd.DataFrame(filter_dataframe(df12))
selected_df
```

```python
# Q10
sales_data = {'Sales': [100, 150, 200, 120, 180, 250, 300, 220, 170, 190],
              'Date': ['2023-06-01', '2023-06-02', '2023-06-03', '2023-06-04', '2023-06-05', '2023-06-06',
                       '2023-06-07', '2023-06-08', '2023-06-09', '2023-06-10']}
df = pd.DataFrame(sales_data)

df['Date'] = pd.to_datetime(df['Date'])
#The min_periods=1 parameter ensures that the moving average is calculated even if there are
# fewer than 7 days of data available
df['moving_average'] = df['Sales'].rolling(window=7,min_periods=1).mean()
df
```

```python
#Q12
import pandas as pd

def select_rows_between_dates(df):
   
     # Select rows between '2023-01-01' and '2023-01-31'
    mask = (df['Date'] >= '2023-01-01') & (df['Date'] <= '2023-01-31')
    selected_rows = df[mask]
    return selected_rows

 
```

```python
select_rows_between_dates(df2)
```
